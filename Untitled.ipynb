```python
# Proyojoniyo library import kora
import pandas as pd

# CSV file-er path define kora
file_path = 'startup data.csv'

# CSV file-ti ke ekta DataFrame-e load kora
try:
    df = pd.read_csv(file_path)

    print("--- Prothom 5 ti row ---")
    print(df.head())
    print("\n" + "="*50 + "\n")

    print("--- Column-er Information ---")
    df.info()
    print("\n" + "="*50 + "\n")

    print("--- Prottek Column-e koto gulo Missing Value ache ---")
    print(df.isnull().sum())
    print("\n" + "="*50 + "\n")
    
    print("--- Company-r Status (Success/Failure) ---")
    print(df['status'].value_counts())

except FileNotFoundError:
    print(f"Error: '{file_path}' file-ti paoya jacche na.")
```

```python
# --- Ei cell-ti ager cell-er por add korun ---

# Prothome, DataFrame (df) theke odorkari column gulo fele di (drop kori)
columns_to_drop = ['Unnamed: 0', 'Unnamed: 6', 'id', 'object_id', 'zip_code', 'latitude', 'longitude', 'name', 'labels', 'city']
cleaned_df = df.drop(columns=columns_to_drop, errors='ignore')

# --- Step 1: Target Variable ('is_success') Toiri Kora ---
# 'status' column-take 1 (success) ebong 0 (failure) te convert kora
# Amra dhorchi 'acquired' ebong 'operating' mane success
# NOTE: apnar output-e 'operating' nei, kintu thakte pare, tai add kora holo
successful_statuses = ['acquired', 'operating']
cleaned_df['is_success'] = cleaned_df['status'].isin(successful_statuses).astype(int)

# Ekhon amader purono 'status' column-er dorkar nei
cleaned_df = cleaned_df.drop('status', axis=1)

# --- Step 2: Missing Data Handle Kora ---
# Je shob row-te kono data nei, shegulo ke fele dewa
# Ekhonkar jonno eti shobcheye shohoj upay
cleaned_df.dropna(inplace=True)

# --- Step 3: Text Column-guloke Number-e Convert Kora ---
# 'state_code' column-ti ekhono text. Eke number-e convert korte hobe.
final_df = pd.get_dummies(cleaned_df, columns=['state_code'], drop_first=True)

# Shesh porjonto, amader toiri kora final DataFrame-ti dekhi
print("--- Data Cleaning Complete! Final Data Head: ---")
print(final_df.head())

print("\n--- Final Data Shape (Rows, Columns) ---")
print(final_df.shape)
```
